Inicializa o dataframe dos dados do Inep curso superior de 2016.

Estamos utilizando chunksize no método `pd.read_csv` já que o dataset é muito grande. Também foi necessário usar o `encoding` e também o `delimiter` com pipe.

Após isso pegamos as primeiras 10000 observações (linhas).

```python
import pandas as pd

curso_superior_chunks = pd.read_csv('DM_ALUNO.CSV', delimiter='|', chunksize=1000, encoding='ISO-8859-1')
```

```python
primeiro_chunk = curso_superior_chunks.get_chunk(10000)
primeiro_chunk
```

```python
primeiro_chunk.describe()
```

```python
# Standard Deviation
primeiro_chunk['CO_COR_RACA_ALUNO'].std()
```

```python
# Observações com valores nulos ou vazio
primeiro_chunk['CO_COR_RACA_ALUNO'].count()
```
