## EDA com Dask

```python
import dask.dataframe as dd
```

```python
inep_2015_data_frame = dd.read_csv('DATA/DM_ALUNO_2015.CSV', delimiter='|', encoding='ISO-8859-1', dtype={'CO_GRAU_ACADEMICO': 'float64',
       'CO_OCDE_AREA_DETALHADA': 'float64',
       'CO_OCDE_AREA_ESPECIFICA': 'float64',
       'CO_OCDE_AREA_GERAL': 'float64',
       'CO_TURNO_ALUNO': 'float64',
       'IN_ING_CONVENIO_PECG': 'float64'})
```
